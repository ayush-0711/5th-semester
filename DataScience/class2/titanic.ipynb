### DATA TRANSFORMATION: Changing the data value for an attribute

```python
df["Survived"]=df["Survived"].map({0:"Died", 1:"Survived"})
df.head()
```

```python
df["Pclass"]=df["Pclass"].map({1:"First", 2:"Second",3:"Third"})
df.head()
```

```python
df["Embarked"]=df["Embarked"].map({'Q':"Queenstown", 'S' : "Southampton", 'C' : "Cherbourg"})
df.head()
```

### Fill NaN Values

```python
# empty embarked -> S
df["Embarked"]=df["Embarked"].fillna("S")
df.head()
```

```python
# empty embarked -> S
df["Cabin"]=df["Cabin"].fillna("XXX")
df.head(5)
```
